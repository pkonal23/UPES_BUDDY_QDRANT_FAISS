```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import DirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os
import glob
```

```python
loader = DirectoryLoader('/Users/konalsmac/upesbuddy/UPESBuddy_Chatbot/web_crawler/summOutput', glob="**/*.txt")
```

```python
docs = loader.load()
```

```python
docs[0]
```

```python
# Import required libraries
import os
from dotenv import load_dotenv

# Load the environment variables from the .env file
load_dotenv()  # This will automatically look for a .env file in the current directory

# Now access the API keys from the environment
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')


# Check if the keys are set correctly (optional)
print(f"OPENAI API Key: {os.environ['OPENAI_API_KEY']}")

```

```python
embeddings = OpenAIEmbeddings(
    model = "text-embedding-3-small"
)
```

```python
text_splitter = RecursiveCharacterTextSplitter()

split_docs = text_splitter.split_documents(docs)
```

```python
split_docs[0]
```

```python
len(split_docs)
```

```python
from langchain_community.vectorstores import Qdrant
```

```python
url_localhost = "http://localhost:6333"
qdrant_docker = Qdrant.from_documents(
    split_docs,
    embeddings,
    url=url_localhost,
    prefer_grpc=False,
    collection_name="upes_scraped_data",
)
```

```python
found_docs_docker = qdrant_docker.similarity_search("who is sanjeev kumar?")
print(found_docs_docker[0].page_content)
```
